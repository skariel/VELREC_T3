```julia
pos = SharedArray(Float32, (3,1))
pos[1,:] = 501
pos[2,:] = 501
pos[3,:] = 501
m = SharedArray(Float32, 1)
m[:] = ones(Float32, 1)
N = 256
rho=SharedArray(Complex64,(N,N,N))
to_cic!(pos,m, rho, 1, 1000);
```

```julia
Winston.set_default_plot_size(400,400)
imagesc(real(rho[:,:,129]))
grid(true)
```

```julia
pos = SharedArray(Float32, (3,150000000))
pos[1,:] = rand(Float32, 150000000)*999+1;
pos[2,:] = rand(Float32, 150000000)*999+1;
pos[3,:] = rand(Float32, 150000000)*999+1;
c = SharedArray(eltype(rho), 150000000)
c[:] = 0;
```

```julia
@time from_cic!(c,pos,rho,1,1000);
```

```julia
mean(pos[1,find(c)])
```

```julia
@time g = to_g_fft!(rho, 1000, 3.0)

imagesc(log(abs(g[:,:,129]))*100)
grid(true)
```

```julia
pos = SharedArray(Float32, (3,1000))
pos[1,:] = ones(1000)*501
pos[2,:] = linspace(503.5,1500,1000)
pos[3,:] = ones(1000)*501
v = SharedArray(eltype(g), 1000)
v[:] = 0

hold(false)
plot(log10(pos[2,:]-501), log10(abs(real(from_cic_dim!(v,pos,g,1,1000,2)))),"k-")
hold(true)

plot(log10(10:999), -2log10(10:999),"g")
hold(false)
grid(true)
```
